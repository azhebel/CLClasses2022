# Задание 1

Составьте с помощью WordNet список существительных, которые могут выступать в качестве прямого дополнения при глаголе to eat, через поиск общего гиперонима для слов meat и potatoes.

```python
meat = wn.synset("meat.n.01")
potato = wn.synset("potato.n.01")

print(meat.hypernyms())
print(potato.hypernyms())

common = meat.lowest_common_hypernyms(potato)
hypo = lambda s: s.hyponyms()
for lemma in list(common[0].closure(hypo)):
    for word in lemma.lemma_names():
        print(word)

```

#   Задание 2

Найдите 10 ближайших синсетов к синсету ‘cow.n.01’.

```python
import collections
```

```python
cow = wn.synset("cow.n.01")

print(cow.definition())

closest = {}

for ss in list(wn.all_synsets(pos=wn.NOUN)):
    # print(ss.name(), cow.path_similarity(ss))
    closest[ss] = cow.path_similarity(ss)

for closest_ss in sorted(closest.items(), key=lambda item: item[1], reverse=True)[:10]:
    print(closest_ss[0], closest_ss[1])
```

# Задание 4

Найдите все ядерные элементы фрейма Removing.

```python
from nltk.corpus import framenet as fn

frame = fn.frame("Removing")
fe = frame.FE

for element in fe:
    if fe[element].coreType == "Core":
        print(element)
```
